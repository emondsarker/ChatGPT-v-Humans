### The distribution of top unigrams before removing stop words in StackOverflow answers 

```python
def get_top_n_words(corpus, n=None):
    vec = CountVectorizer().fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]
common_words = get_top_n_words(df['aa_body'], 10)
for word, freq in common_words:
    print(word, freq)
df1 = pd.DataFrame(common_words, columns = ['aa_body_top' , 'count'])
df1.groupby('aa_body_top').sum()['count'].sort_values(ascending=False).iplot(
    kind='bar', yTitle='Count', linecolor='black', title='Top 100 words in StackOverflow answers before removing stop words')
```

### The distribution of top unigrams after removing stop words

```python
def get_top_n_words(corpus, n=None):
    vec = CountVectorizer(stop_words = 'english').fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]
common_words = get_top_n_words(df['aa_body'], 10)
for word, freq in common_words:
    print(word, freq)
df2 = pd.DataFrame(common_words, columns = ['aa_body_top' , 'count'])
df2.groupby('aa_body_top').sum()['count'].sort_values(ascending=False).iplot(
    kind='bar', yTitle='Count', linecolor='black', title='Top 100 words in StackOverflow answers after removing stop words')
```

### The distribution of top bigrams before removing stop words

### The distribution of top part-of-speech tags of ans corpus

```python
blob = TextBlob(str(df['aa_body']))
pos_df = pd.DataFrame(blob.tags, columns = ['word' , 'pos'])
pos_df = pos_df.pos.value_counts()[:20]
pos_df.iplot(
    kind='bar',
    xTitle='POS',
    yTitle='count', 
    title='Top 20 Part-of-speech tagging for answer corpus')
```
