```python
import pandas as pd
import string
```

```python
df = pd.read_csv('/content/drive/MyDrive/ChatGPTvHumans/its embedding time/stack_cgpt_embedding.xlsx - Sheet1.csv')
```

```python
df = df.astype(str)
df.head()
```

```python
stack_cgpt = df['stack_cgpt']
stack_human = df['stack_human']
```

```python
# clean each row and make a list of keywords
arr1 = []
arr2 = []
for text in stack_cgpt:
  if text != "nan":
    text = text.translate(str.maketrans('','',string.punctuation))
    arr1.append(text)

for text in stack_human:
  if text != "nan":
    text = text.translate(str.maketrans('','',string.punctuation))
    arr2.append(text)                               
```

```python
print(len(arr1),len(arr2))
```

```python
# create sets of each array
gpt_set = set(arr1)
human_set = set(arr2)
```

```python
# find the words that are different in each set
difference = gpt_set.symmetric_difference(human_set)
```

```python
print(len(difference))
```

```python
# turn back to list
difference = list(difference)
```

```python
difference
```

```python
# read the two csv you want to clean
df1 = pd.read_csv('/content/drive/MyDrive/ChatGPTvHumans/its embedding time/stack_cgpt_embedding.csv', header=None)
df2 = pd.read_csv('/content/drive/MyDrive/ChatGPTvHumans/its embedding time/stack_human_embedding.csv', header=None)
```

```python
df1.head()
```

```python
# function to remove punctuations
def remove_punc(text):
  return str(text).translate(str.maketrans('','',string.punctuation))
```

```python
# apply function to column 0
df1.iloc[:,0] = df1.iloc[:,0].apply(lambda x: remove_punc(x))
df2.iloc[:,0] = df2.iloc[:,0].apply(lambda x: remove_punc(x))
```

```python
# remove values that are different
df1 = df1[~df1.iloc[:,0].isin(difference)]
df2 = df2[~df2.iloc[:,0].isin(difference)]
```

```python
# drop null values
df1 = df1[df1.iloc[:,0]!=""]
df2 = df2[df2.iloc[:,0]!=""]
```

```python
# drop duplicate rows by column 0
df1 = df1.drop_duplicates(subset=[0])
df2 = df2.drop_duplicates(subset=[0])
```

```python
# check size
print(df1.size, df2.size)
```

```python
# save files
df1.to_csv('/content/df1.csv')
df2.to_csv('/content/df2.csv')
```
