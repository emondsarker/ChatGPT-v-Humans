```python
import pandas as pd
from bs4 import BeautifulSoup
import os
```

```python
df1 = pd.read_csv('/content/drive/MyDrive/ChatGPTvHumans/StackOverFlow.csv')
df1.head()
```

```python
df2 = pd.DataFrame(columns=['Title','q_body','aa_body'])
df2.head()
```

```python
print(df1.loc[0,['q_body']].q_body)
```

```python
# loop through dataframe 1
for index in range(len(df1)):


  # get question & answer from df1
  question = df1.loc[index,['q_body']].q_body
  answer = df1.loc[index,['aa_body']].aa_body

  # soupify
  question_soup = BeautifulSoup(question, 'html.parser')
  answer_soup = BeautifulSoup(answer, 'html.parser')

  # get text
  clean_question = question_soup.get_text()
  clean_answer = answer_soup.get_text()

  # store to df2
  df2.loc[index,['Title']] = df1.loc[index,['Title']].Title
  df2.loc[index,['q_body']] = clean_question.replace("\n","")
  df2.loc[index,['aa_body']] = clean_answer.replace("\n","")

df2.head()
```

```python
df2.to_csv('/content/drive/MyDrive/ChatGPTvHumans/cleaned_stackoverflow.csv')
```
